# 1. Строка с индксом 19      Нужен ли здесь inplace=True????????

К настоящему моменту мы уже знаем, как находить среднее арифметическое и среднее квадратическое отклонение. Проблема же этих метрик, как мы уже говорили, заключается в том, что они сильно подвержены выбросам.
Медианой называется число, которое находится в середине упорядоченного от меньшего к большему набора чисел.
Когда медиана меньше среднего арифметического, мы наблюдаем скошенное вправо распределение.
В этом смысле медиана дает более надежную оценку среднего при наличии выбросов.

```python
# Заменим значения в столбцах 'Шаг нашивки' и  'Плотность нашивки' со значения "0.0" на медиану значений в этих столбцах

hw_data_composite_add.loc[19, 'Шаг нашивки'] = hw_data_composite_add['Шаг нашивки'].median()
hw_data_composite_add.loc[19, 'Шаг нашивки'] # значение в точке строка с индексом 19 и столбец 'Шаг нашивки'
```

```python
hw_data_composite_add.loc[19, 'Плотность нашивки'] = hw_data_composite_add['Плотность нашивки'].median()
hw_data_composite_add.loc[19, 'Плотность нашивки'] # значение в точке строка с индексом 19 и столбец 'Плотность нашивки'
```

```python
# Проверим произошла ли замена нулевых значений на мадианы в столбцах 'Шаг нашивки' и  'Плотность нашивки'
hw_data_composite_add.iloc[[19]] 
```

# да, замена произведена
