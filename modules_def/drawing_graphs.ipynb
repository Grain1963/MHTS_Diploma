```python
# Построение графиков

# Функция drawing_graphs построения основных графиков по парметрам df:  sns.histplot и sns.boxplot, sns.kdeplot, sns.scatterplot и sns.displot
# Параметры, передаваемые в функцию:
# df- наш dataframe
# plt_type - тип графика ,который выводится (histplot, boxplot, displot, kdeplot, catterplot)
# n_col - количество столбцов графиков сколько будем строить в строке
# figsize_a - размер по ширине
# figsize_b - размер по высоте
# name_Y - наименование параметра для графика типа regplot
# plt_sutitle - 'Общее назвавние серии графиков '
# graf_name_path - путь записи в директорию fig_block1_clean имени рисунка graf_df_bp_40'  -пример -   r'save_fig\fig_block1_clean\graf_df_bp_40'

def drawing_graphs(df, n_row, n_col, figsize_a, figsize_b, name_Y, plt_type, plt_suptitle, graf_name_path):
    plt.figure(figsize=(figsize_a, figsize_b)) 
    a = n_row
    b = n_col
    c = 1
    plt.suptitle(plt_suptitle) # Печать заголовка
    
    for column in df.columns:
        plt.subplot(a, b, c)

        if  plt_type == 'histplot':
            sns.histplot(df[column],
                         bins = 30, 
                         kde = True, # оценка плотности ядра (kernel density estimation или сокращенно kde)
                         fill = True
                         )    
        elif plt_type == 'boxplot':       
            sns.boxplot(data=df, 
                        x=column,
                        )
        elif plt_type == 'kdeplot':
             sns.kdeplot(df[column], 
                         bw_adjust= 1
                        ) 
        elif plt_type == 'displot':
             sns.displot(data =df, 
                         x = column, 
                         kind = 'kde',
                         fill = True
                         )
        elif plt_type == 'scatterplot':
            sns.scatterplot(df[column],
                            alpha=1,
                            color='b'
                            )
        elif plt_type == 'regplot':    
            sns.regplot(data = df,
                        x = df[column],
                        y = name_Y,
                        ci= None,
                        ) # ci=None- без доверительного интервала 
        elif plt_type == 'catplot':
            if column == 'Model':
                continue
            sns.catplot(df, kind='bar',x= 'Model', y= column,
                        height=3, aspect=2) # 
            plt.xticks(size=5)
        
        plt.ylabel(None)    
        plt.title(column, size = 10)
        plt.tight_layout() # название осей не будут наезжать на соседние графики
        plt.savefig(graf_name_path, dpi=100) # Сохраняем рисунок таблицы в папку save_fig
        c += 1
        
    return
```
